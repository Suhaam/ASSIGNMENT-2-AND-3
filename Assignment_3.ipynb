```python
#Visualize the no. of employess with ditance from home by attrition
import matplotlib.pyplot as plt
plt.subplots(figsize=(18,10))
sns.countplot(x= 'DistanceFromHome', hue = 'Attrition', data = df, palette = 'colorblind')
```

```python
%matplotlib inline

#VISULAIZE question 1
#Breakdown of distance from home by job role and attrition.

df_DistanceFromHome=df.groupby(["JobRole", "Attrition"])["DistanceFromHome"].sum().reset_index()
plt.figure(figsize=(20,10))
sns.barplot(x="JobRole", y="DistanceFromHome", hue="Attrition", data = df_DistanceFromHome)
plt.title("Breakdown of the Distance from home by BOTH(job role and attrition)")
plt.xlabel("JobRole")
plt.ylabel("DistanceFromHome")
```

```python
##VISULAIZING question 2
#Compare average monthly income by education and attrition.

%matplotlib inline

df_MonthlyIncome=df.groupby(["Education", "Attrition"])["MonthlyIncome"].sum().reset_index()
plt.figure(figsize=(20,10))
sns.barplot(x = "Education", y = "MonthlyIncome", hue = "Attrition", data = df_MonthlyIncome)
plt.title("Comparing the MonthlyIncome by Education and Attrition")
plt.xlabel("Education")
plt.ylabel("MonthlyIncome")

```
